```python
x = tf.constant(2)
y = tf.constant(3)
```

```python
print('Operations with Constants')
print('Addition:', (x + y).numpy())
print('Subtraction:', (x - y).numpy())
print('Multiplication:', (x * y).numpy())
print('Division:', (x / y).numpy())
```

```python
x = tf.constant(10, dtype=tf.int32)
y = tf.constant(5, dtype=tf.int32)
```

```python
add = tf.add(x, y)
sub = tf.subtract(x, y)
mul = tf.multiply(x, y)
```

```python
print("Operations with Constants")
print("Addition:", add.numpy())
print("Subtraction:", sub.numpy())
print("Multiplication:", mul.numpy())
```
