```python
class circle():
    pi = 3.14

    def __init__(self, radius =1):
        self.radius = radius
        self.area = radius*radius*circle.pi
        
    def get_circumference(self):
        return self.radius * circle.pi * 2
```

```python
my_circle = circle(30)
```

```python
my_circle.pi
```

```python
my_circle.radius
```

```python
my_circle.get_circumference()
```

```python
my_circle.area
```

```python
class Animal:

    def __init__(self):
        print("Animal Created!")

    def who_am_i(self):
        print("I am an animal")

    def eat(self):
        print("I am eating")
              
```

```python
my_animal = Animal()
```

```python
my_animal.eat()
```

```python
my_animal.who_am_i()
```

```python
class Dog(Animal):

    def __init__(self):
        Animal.__init__(self)
        print("Dog created!!")

    def eat(self):
        print("I am a dog and eating")
        
    def bark(self):
        print("WOOF!")
```

```python
mydog = Dog()
```

```python
mydog.eat()
```

```python
mydog.who_am_i()
```

```python
class Dog():

    def __init__(self,name):
        self.name = name

    def speak(self):
        return self.name + " says woof!!"
```

```python
class Book():

    def __init__(self,title,author,pages):
        self.title = title
        self.author = author
        self.pages = pages

    def __str__(self):
        return f"{self.title} by {self.author}"

    def __len__(self):
        return self.pages

    def __del__(self):
        print("A book object has been deleted!!")
```

```python
b = Book('Python Rocks', 'Harsh', 200)
```

```python
print(b)
```

```python
str(b)
```

```python
len(b)
```

```python
del b
```
