```python
import PyPDF2
```

```python
f = open('Find_the_Phone_Number.pdf','rb')
pdf = PyPDF2.PdfReader(f)
len(pdf.pages)
```

```python
import re
```

```python
pattern = r'\d{3}.\d{3}.\d{4}' 
for n in range(len(pdf.pages)):
    
    page  = pdf.pages[n]
    page_text = page.extract_text()
    match = re.search(pattern,page_text)
    
    if match:
        print(match.group())
```
