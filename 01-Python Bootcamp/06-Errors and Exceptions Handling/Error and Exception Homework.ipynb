```python
try:
    for i in ['a', 'b', 'c']:
        print(i**2)
except TypeError:
    print("You have an type error")
except:
    print("You have an error")
finally:
    print("I always run")
```

```python
try:
    x=5
    y=0
    z=x/y
except ZeroDivisionError:
    print("Division by zero is not allowed")
except:
    print("You have an error")
finally:
    print("All Done")
```

```python
def ask():
    num=int(input("Enter a number: "))
    print(f"Square of {num} is {num**2}")
```

```python
while True:
    try:
        ask()
    except TypeError:
        print("You have an type error")
        continue
    except:
        print("You have an error")
        continue
    else:
        print("Yes thank you!")
        break
    finally:
        print("I am finally block")
```
