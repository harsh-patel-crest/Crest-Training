```python
def gensquares(n):
    for i in range(n):
        yield i**2
```

```python
for x in gensquares(10):
    print(x)
```

```python
def rand_num(low,high,n):
    for i in range(n):
        num=random.randint(low,high)
        yield num
```

```python
for num in rand_num(1,10,12):
    print(num)
```
