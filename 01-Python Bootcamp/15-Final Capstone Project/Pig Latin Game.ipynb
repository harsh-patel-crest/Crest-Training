```python
vowels=['a','e','i','o','u']
consonants=['b','c','d','f','g','h','j','k','l','m','n','p','q','r','s','t','v','w','x','y','z']
```

```python
def pig_latin(s):
    if not s:
        return "Empty String"
    if s[0].lower() in vowels:
        word=s+'-ay'
    else:
        i=0
        while s[i].lower() in consonants:
            i+=1
        word=s[i:]+'-'+s[:i]+'ay'
    return word
```

```python
s=input("Enter a string to create a Pig Latin Game: ")
print(f"Pig Latin Word: {pig_latin(s)}")
```
