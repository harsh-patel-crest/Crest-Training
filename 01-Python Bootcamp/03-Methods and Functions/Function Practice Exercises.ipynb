```python
def lesser_of_two_even(a,b):
    if(a%2==0 and b%2==0):
        if a<b:
            return a
        else:
            return b
    else:
        if a>b:
            return a
        else:
            return b
        
```

```python
lesser_of_two_even(2,4)
```

```python
lesser_of_two_even(2,5)
```

```python
def animal_crackers(text):
    mylist=[text[0] for text in text.split()]
    if mylist[0]==mylist[1]:
        return True
    else:
        return False
```

```python
def makes_twenty(a,b):
    if ((a==20 or b==20) or a+b==20):
        return True
    else:
        return False
```

```python
def old_macdonald(name):
    if len(name) < 4:
        return 'name is too short!!'
    return name[0].upper() + name[1:3] + name[3].upper() + name[4:]
```

```python
old_macdonald('harsh')
```

```python
def master_yoda(text):
    words = text.split()
    words.reverse()
    return " ".join(words)
```

```python
master_yoda('I am home')
```

```python
master_yoda('We are ready')
```

```python
def has_33(nums):
    for i in range(len(nums)-1):
        if nums[i] == nums[i+1] and nums[i]==3:
            return True
    return False
```

```python
has_33([1, 3, 3])
```

```python
has_33([1, 3, 1, 3])
```

```python
has_33([3, 1, 3])
```

```python
def blackjack(a,b,c):
    if sum([a,b,c])<=21:
        return sum([a,b,c])
    elif sum([a,b,c])>21 and (a==11 or b==11 or c==11):
        return sum([a,b,c])-10
    else:
        return 'BUST'
```

```python
blackjack(5,6,7)
```

```python
blackjack(9,9,9)
```

```python
blackjack(9,9,11)
```

```python
def summer_69(arr):
    total = 0
    add = True

    for num in arr:
        if add:
            if num == 6:
                add = False
            else:
                total += num

        else:
            if num == 9:
                add = True

    return total
```

```python
def spy_game(nums):
    code = [0, 0, 7]
    for num in nums:
        if num == code[0]:
            code.pop(0)
        if not code:
            return True
    return False
```

```python
spy_game([1,2,4,0,0,7,5])
```

```python
spy_game([1,0,2,4,0,5,7])
```

```python
spy_game([1,7,2,0,4,5,0])
```

```python
def count_primes(num):
    if num<2:
        return 0

    primes=[2]

    x=3
    while x<=num:
        for y in range(3,x,2):
            if x%y==0:
                x+=2
                break
        else:
            primes.append(x)
            x+=2
    print(primes)
    return len(primes)
```

```python
count_primes(100)
```
