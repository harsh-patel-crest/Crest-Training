```python
def ran_check(num,low,high):
    return num in range(low,high+1)
```

```python
ran_check(5,2,7)
```

```python
def up_low(s):
    upper=0
    lower=0
    for char in s:
        if char.isupper():
            upper+=1
        elif char.islower():
            lower+=1
    return upper, lower
```

```python
up_low("Hiii I am Harsh Patel")
```

```python
def unique_list(lst):
    return list(set(lst))
```

```python
unique_list([1,2,3,4,5,1,1,1,5,5,5,2,2,2,4,3,3])
```

```python
def multiply(numbers):
    result=1
    for num in numbers:
        result*=num
    return result
```

```python
multiply([1,2,3,-4,5])
```

```python
def palindrome(s):
    s=s.replace(" ", "")
    return s==s[::-1]
```

```python
palindrome('nurses run')
```

```python
import string
def ispangram(str1, alphabet=string.ascii_lowercase):
    alphaset=set(alphabet)
    str1=str1.replace(' ','')
    str1=str1.lower()
    str1=set(str1)
    return alphaset==str1
```

```python
ispangram("The quick brown fox jumps over the lazy dog")
```

```python
ispangram('Harsh Patel')
```
