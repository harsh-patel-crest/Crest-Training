```python
row1=[' ',' ',' ']
row2=[' ',' ',' ']
row3=[' ',' ',' ']
```

```python
def display(row1,row2,row3):
    print(row1)
    print(row2)
    print(row3)
```

```python
result=''
```

```python
def update_board(row, index, value):
    global result
    if row==1 and row1[index]==' ':
        row1[index]=value
    elif row==2 and row2[index]==' ':
        row2[index]=value
    elif row==3 and row3[index]==' ':
        row3[index]=value
    else:
        print("Invalid Row Number or Position!")
    display(row1,row2,row3)
    result = check_win(row1,row2,row3)
```

```python
def check_win(row1,row2,row3):
    if row1==['X']*3 or row1==['O']*3:
        return 'You Win!'
    elif row2==['X']*3 or row2==['O']*3:
        return 'You Win!'
    elif row3==['X']*3 or row3==['O']*3:
        return 'You Win!'
    elif (row1[0]=='X' or row1[0]=='O') and row1[0]==row2[1]==row3[2]:
        return 'You Win!'
    elif (row1[2]=='X' or row1[2]=='O') and row1[2]==row2[1]==row3[0]:
        return 'You Win!'
    elif (row1[0]=='X' or row1[0]=='O') and row1[0]==row2[0]==row3[0]:
        return 'You Win!'
    elif (row1[1]=='X' or row1[1]=='O') and row1[1]==row2[1]==row3[1]:
        return 'You Win!'
    elif (row1[2]=='X' or row1[2]=='O') and row1[2]==row2[2]==row3[2]:
        return 'You Win!'
```

```python
def user_input():
    row=int(input(("Enter Row Number (1-3):")))
    index=int(input("Enter Index Position (0-2):"))
    value=input("Enter Value (O or X):")
    update_board(row,index,value)
```
