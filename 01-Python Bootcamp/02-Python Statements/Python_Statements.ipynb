## While loops in python

````python
x = 0

while x < 5:
    print(f'The current value of x is {x}')
    x = x+1
````

### break, continue, pass

we can use break, continue, and pass statements in our loops to add additinonal functionality

break: Breaks out of the current closest enclosing loop.<br>
continue: Goes to the top of the closest enclosing loop.<br>
pass: Does nothing at all.

````python
x = [1,2,3]

for item in x:

    pass


print('end of script')
````

````python
mystring = 'Sammy'

for letter in mystring:
    if letter == 'a':
        break;
    print(letter)
````

````python
x = 0

while x<5:
    print(x)
    x = x+1
````

# List Comprehension

````python
mystring='hello'
````

````python
mylist=[]

for letter in mystring:
    mylist.append(letter)
````

````python
mylist
````

````python
mylist=[letter for letter in mylist]
````

````python
mylist
````

````python
mylist=[num**2 for num in range(0,11)]
````

````python
mylist
````

````python
celcius=[0,10,20,34.5]

fahrenheit=[((9/5)*temp+32) for temp in celcius]
````
