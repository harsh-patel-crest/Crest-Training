```python
import random
guesses = []

number = random.randint(1,100)
```

```python
while True:
    try:
        guess = int(input("number:"))
    except:
        print("Enter valid num")
        continue

    if guess < 1 or guess > 100:
        print("OUT OF BOUNDS")
        continue

    guesses.append(guess)

    if guess == number:
        print("Matched!!")
        break

    if len(guesses) == 1:
        if abs(guess - number) <= 10:
            print("WARM!")
        else:
            print("COLD!")
    else:
        if abs(guess - number) < abs(guesses[-2] - number):
            print("WARMER!")
        else:
            print("COLDER!")
```
