```python
st='Print only the words that start with s in this sentence'
```

```python
st1=st.split()

for letter in st1:
    if letter[0] == 's':
        print(letter)
```

```python
st='Print every word in this sentence that has an even number of letters'

st = st.split()
for l in st:
    if len(l)%2==0:
        print("Even!")
        print(l)
```

```python
for num in range(1,101):
    
    if num%3==0 and num%5==0:
        print("FizzBuzz")
    elif num%3==0:
        print("Fizz")
    elif (num%5==0):
        print("Buzz")
    else:
        print(num)
```
