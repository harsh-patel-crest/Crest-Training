```python
from random import randint

rnum=randint(1,100)
```

```python
unum=None
while unum!=rnum:
    unum = int(input("Enter a number:"))
    if unum>rnum:
        print("Number is smaller than expected")
    elif unum<rnum:
        print("Number is greater than expected")
    else:
        print("You guessed correctly!")
```
