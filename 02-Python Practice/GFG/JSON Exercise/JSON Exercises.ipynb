```python
import json
from collections import namedtuple
```

```python
data = '{"name" : "Geek", "id" :1, "location": "Mumbai"}'

x = json.loads(data, object_hook = lambda d: namedtuple('X', d.keys())(*d.values()))
```

```python
unflat_json = {'user':
               {'Harsh':
                {'UserID': 1717171717,
                 'Email': 'harsh2004@gmail.com',
                 'friends': ['Piyush', 'Dhyan', 'Aayushi']
                 }
                }
               }
```

```python
def flatten_json(y):
    out = {}

    def flatten(x, name=''):

        if type(x) is dict:

            for a in x:
                flatten(x[a], name + a + '_')
                
        elif type(x) is list:

            i = 0

            for a in x:
                flatten(a, name + str(i) + '_')
                i += 1
        else:
            out[name[:-1]] = x

    flatten(y)
    return out
```

```python
print(flatten_json(unflat_json))
```
