```python
from math import pi

def pi_upto_n_dec(dec):
    return float(f"{pi:.{dec}f}")

pi_upto_n_dec(10)
```

```python
import random

r=random.randint(0,100)
flag=0
while flag==0:
    num=int(input("Enter a number:"))
    if num<r:
        print("You are small!")
    elif num>r:
        print("You are big!")
    else:
        flag=1
        print("You are perfect!")
```
