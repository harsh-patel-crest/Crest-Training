```python
def is_even(num):
    return num%2==0

num=int(input("Enter to number:"))
is_even(num)
if is_even(num)==True:
    print(f"{num} is even")
else:
    print(f"{num} is odd")
```

```python
def bool_operations(b1,b2):
    return ((b1 and b2), (b1 or b2))
```

```python
b1=input("Enter True of False:")
b2=input("Enter True of False:")
if b1=="true" or b1=="TRUE" or b1=="True":
    b1=True
else:
    b1=False

if b2=="true" or b2=="TRUE" or b2=="True":
    b2=True
else:
    b2=False

print(bool_operations(b1,b2))
```

```python
def vote_eligible(age):
    return age>=18
```

```python
age=int(input("Enter your age:"))
vote_eligible(age)
if vote_eligible(age):
    print("You are eligible for voting!")
else:
    print("You are not eligible!")
```
