```python
from collections import Counter

s='Python Exercise!'
count=Counter(s)
print(count)
```

```python
s='Python Exercise!'
vowels=['a','e','i','o','u','A','E','I','O','U']

count=Counter(c for c in s if c in vowels)

print(count)
```
