```python
def main():
    x = frozenset([1, 2, 3, 4, 5])
    y = frozenset([0, 1, 3, 7, 8, 10])    
    print(x)
    print(y)
    union_result = x.union(y)
    print("Union:", union_result)
    intersection_result = x.intersection(y)
    print("Intersection:", intersection_result)
    difference_result1 = x.difference(y)
    difference_result2 = y.difference(x)
    print("Difference of x and y", difference_result1)
    print("Difference of y and x", difference_result2)

if __name__ == "__main__":
    main()
```

```python
f_set=frozenset([1,2,3,4,5])
r_set=set([3,4,5,6,7])

convert_to_set=set(f_set)
convert_to_frozenset=frozenset(r_set)

print(f_set)
print(type(f_set))
print(r_set)
print(type(r_set))
print(convert_to_set)
print(type(convert_to_set))
print(convert_to_frozenset)
print(type(convert_to_frozenset))
```
