```python
from collections import namedtuple

Point=namedtuple('Point',['x','y'])

p=Point(11,22)

print("x:",p.x)
print("y:",p.y)

print("x:",p[0])
print("y:",p[1])
```

```python
Employee=namedtuple('Employee',['name','age','country'])

employees=[Employee("Harsh Patel", 21, "India"),Employee("Het Patel", 20, "Australia"),Employee("Jansi Patel", 21, "USA"),Employee("Kesha Patel", 21, "Canada")]

for employee in employees:
    print("Employee Name:", employee.name)
    print("Employee Country:", employee.country)
```

```python
Food=namedtuple('Food',['name', 'price'])
food1=Food('Pizza',100)
food2=Food('Samosa',20)
food3=Food('Burger',50)

print(f'Price of {food3.name} is {food3.price}')
```
