```python
from collections import OrderedDict

o_dict=OrderedDict()

o_dict['Laptop']=15
o_dict['Desktop']=40
o_dict['Mobile']=50

for item,quantity in o_dict.items():
    print(f"Item:{item}, Quantity:{quantity}")
```

```python
sort_by_key=OrderedDict(sorted(o_dict.items(),key=lambda item:item[0]))
sort_by_value=OrderedDict(sorted(o_dict.items(),key=lambda item:item[1]))

print(sort_by_key)
print(sort_by_value)
```

```python
reversed_dict=OrderedDict(reversed(list(o_dict.items())))
print(reversed_dict)
```
