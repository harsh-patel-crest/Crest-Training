```python
def string(s):
    if not s:
        return None
    return s
```

```python
s=input("Enter a string:")
print(string(s))
```

```python
def mid_char(s):
    if len(s)%2==0:
        return None
    return s[int(len(s)/2)]
```

```python
s=input("Enter a string:")
print(mid_char(s))
```

```python
def print_str(l):
    for string in l:
        print(string)
        if not string:
            print("Empty String")
```

```python
l=['Harsh', 'Sunil','', 'Het','']
print_str(l)
```
