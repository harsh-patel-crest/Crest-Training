```python
import numpy as np
```

```python
print("Add:")
print(np.add(1, 4))
print("Subtract:")
print(np.subtract(1, 4))
print("Multiply:")
print(np.multiply(1, 4))
print("Divide:")
print(np.divide(1, 4))
```

```python
x = np.arange(10)
print(x)
print("Division of the array inputs, element-wise:")
print(np.true_divide(x, 3)) 
```

```python
x = [1., 2., 3., 4.]
print(x)
print("Largest integer smaller or equal to the division of the inputs:")
print(np.floor_divide(x, 1.5))
```
