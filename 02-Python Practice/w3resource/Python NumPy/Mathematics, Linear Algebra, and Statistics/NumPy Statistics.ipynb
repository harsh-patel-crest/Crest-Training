```python
import numpy as np
```

```python
x=np.arange(1,10).reshape(3,3)
print(x)

x=x.flatten()
print(x)
```

```python
max_val=np.max(x)
min_val=np.min(x)
print(max_val)
print(min_val)
```

```python
x=np.arange(1,10).reshape(3,3)
print(x)

max_val=np.max(x,axis=1)
min_val=np.min(x,axis=1)
print(max_val)
print(min_val)
```

```python
r=np.ptp(x,axis=1)
print(r)
```
