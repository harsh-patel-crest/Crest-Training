```python
def list_sum(l):
    if not l:
        return 0
    else:
        return list_sum[0] + list_sum(l[1:])
```

```python
l=[1,2,3,4,5]
sum(l)
```

```python
def nested_list_sum(l):
    total = 0
    for element in l:
        if isinstance(element, list):
            total += nested_list_sum(element)
        else:
            total += element
    return total
```

```python
l=[1,2,[3,4],[5,6]]
nested_list_sum(l)
```

```python
def factorial(num):
    if num<0:
        return "Factorial not possible"
    elif num==0:
        return 1
    else:
        return num*factorial(num-1)
```

```python
factorial(6)
```

```python
def fibonacci(num):
    if num<=1:
        return num
    else:
        return fibonacci(num-1)+fibonacci(num-2)
```

```python
for i in range(8):
    print(fibonacci(i))
```

```python
def sum_digits(num):
    if num<=9:
        return num
    else:
        return num%10+sum_digits(int(num/10))
```

```python
sum_digits(345)
sum_digits(45)
```
