```python
import pandas as pd

df = pd.DataFrame({
    'A': [1, 6, 8, 3, 7],
    'B': [5, 2, 9, 4, 1]
})

result = df[df['A'] > 4]
print(result)
```

```python
df = pd.DataFrame({
    'X':[1, 6, 8, 3, 7],
    'Y':[5, 2, 9, 4, 1],
    'Z':[3, 8, 6, 2, 9]
})

result = df[['X','Y']]
print(result)
```

```python
df = pd.DataFrame({
    'X': [1, 6, 8, 3, 7],
    'Y': [5, 2, 9, 4, 1],
    'Z': ['one', 'one', 'two', 'two', 'one']
})

df = df.set_index(['Z', 'X'])

result = df.loc[('one', 6)]
print(result)
```

```python
df = pd.DataFrame({
    'A': [1, 6, 8, 3, 7],
    'B': [5, 2, 9, 4, 1],
    'C': ['one', 'one', 'two', 'two', 'one']
})

df = df.set_index(['C', 'A'])

result = df.loc['one']
print(result)
```
