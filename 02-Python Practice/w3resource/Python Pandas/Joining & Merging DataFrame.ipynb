```python
import pandas as pd

student_data1 = pd.DataFrame({
        'student_id': ['S1', 'S2', 'S3', 'S4', 'S5'],
         'name': ['Harsh', 'Piyush', 'Aayushi', 'Het', 'Kesha'], 
        'marks': [200, 210, 190, 222, 199]})

student_data2 = pd.DataFrame({
        'student_id': ['S4', 'S5', 'S6', 'S7', 'S8'],
        'name': ['Riya', 'Eti', 'Sunil', 'Jansi', 'Dhyan'], 
        'marks': [201, 200, 198, 219, 201]})

print("Original DataFrames:")
print(student_data1)
print("-------------------------------------")
print(student_data2)
print("\nJoin the said two dataframes along rows:")
result_data = pd.concat([student_data1, student_data2])
print(result_data)
```

```python
print("\nJoin the said two dataframes along columns:")
result_data = pd.concat([student_data1, student_data2], axis = 1)
print(result_data)
```

```python
s6 = pd.Series(['S6', 'Daiwik', 205], 
               index=['student_id', 'name', 'marks'])

print("Original DataFrame:")
print(student_data1)

print("\nNew Row:")
print(s6)

combined_data = pd.concat([student_data1, s6.to_frame().T], ignore_index=True)

print("\nCombined Data:")
print(combined_data)
```
