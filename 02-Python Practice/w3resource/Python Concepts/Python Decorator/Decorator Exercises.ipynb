```python
def my_decorator(func):
    def wrapper():
        print("1. [Before] Putting on a coat...")
        func()
        print("2. [After] Taking off the coat.")
    return wrapper

@my_decorator
def go_outside():
    print("   -> I am walking outside.")

go_outside()
```

```python
def decorator(func):
    def wrap(*args, **kwargs):
        print(f"Calling {func.__name__} with args: {args}, kwargs: {kwargs}")
        
        result = func(*args, **kwargs)
        
        print(f"{func.__name__} returned: {result}")
        
        return result
    return wrap

@decorator
def multiply_numbers(x, y):
    return x * y

result = multiply_numbers(10, 20)
print("Result:", result)
```

```python
import time

def decorator(func):
    def wrap(x,y):
        start = time.perf_counter()
        result=func(x,y)
        end = time.perf_counter()
        print(f"Time: {end - start} seconds")
        return result
    return wrap

@decorator
def multiply_numbers(x, y):
    return x * y

result = multiply_numbers(119, 209)
print("Result:", result)
```
