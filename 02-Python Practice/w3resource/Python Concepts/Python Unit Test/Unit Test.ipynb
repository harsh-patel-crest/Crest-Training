```python
def is_prime(num):
    if num<=1:
        return False
    else:
        for i in range(2,int(num**0.5)+1):
            if num%i==0:
                return False
        return True
```

```python
num=int(input("Enter a number to check if it is prime or not:"))
is_prime(num)
```

```python
def is_sorted(l):
    return l==sorted(l)
```
