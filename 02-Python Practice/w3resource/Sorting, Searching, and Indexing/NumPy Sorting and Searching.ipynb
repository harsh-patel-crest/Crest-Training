```python
data_type = [('name', 'U10'), ('class', int), ('height', float)]
students_details = [('Harsh', 5, 48.5), ('Piyush', 6, 52.5), ('Dhyan', 5, 42.10), ('Daiwik', 5, 40.11)]
students = np.array(students_details, dtype=data_type)
print("Original array:")
print(students)
print("Sort by height")
print(np.sort(students, order='height'))   
```

```python
data_type = [('name', 'U10'), ('class', int), ('height', float)]
students_details = [('Harsh', 5, 48.5), ('Piyush', 6, 52.5), ('Dhyan', 5, 42.10), ('Daiwik', 6, 40.11)]
students = np.array(students_details, dtype=data_type)
print("Original array:")
print(students)
print("Sort by class")
print(np.sort(students, order=['class', 'height']))   
```
