```python
import pandas as pd

# Create a sample DataFrame
data = {'Name': ['Harsh', 'Het', 'Piyush', 'Kesha', 'Pratik'],
        'Age': [26, 32, 25, 31, 28],
        'Salary': [50000, 60000, 45000, 70000, 55000]}

df = pd.DataFrame(data)

selected_rows = df[(df['Age'] > 25) & (df['Salary'] > 50000)]

print(selected_rows)
```

```python
selected_rows = pd.concat([df.head(3)])

print("First 3 rows:")
print(selected_rows)

selected_rows = pd.concat([df.tail(3)])
print("\nLast 3 rows:")
print(selected_rows)
```

```python
filtered_rows = df[df['Age'] > 30]

print(filtered_rows)
```

```python
df['Double_Age'] = np.multiply(df['Age'], 2)
print("New Dataframe:")
print(df)
```

```python
selected_age_values = np.array([25, 35])

selected_rows = df[df['Age'].isin(selected_age_values)]

print(selected_rows)
```

```python
numpy_array = np.array([1000, 2000, 3000, 4000, 5000])

df['Updated_Salary'] = np.add(df['Salary'], numpy_array)

print(df)
```
