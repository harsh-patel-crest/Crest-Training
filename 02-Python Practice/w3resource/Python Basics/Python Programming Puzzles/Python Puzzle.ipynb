### Check Nineteen and Five Occurrences

```python
list1=[19,19,15,5,3,5,5,2]

def find_occurences(l):
    count_19=0
    count_5=0
    for i in range(len(l)):
        if l[i]==19:
            count_19+=1
        elif l[i]==5:
            count_5+=1
    if count_19==2 and count_5>=3:
        return True
    else:
        return False
```

```python
print(find_occurences(list1))
```

```python
list2=[19,15,15,5,3,3,5,2]
```

```python
print(find_occurences(list2))
```

```python
list3=[19,19,5,5,5,5,5]
```

```python
print(find_occurences(list3))
```

### Fifth Element and List Length Check

```python
def check(l):
    if len(l)==8 and l.count(l[4])==3:
        return True
    else:
        print(f"Length is {len(l)} and Fifth element is {l[4]}")
        return False
```

```python
l1=[19, 19, 15, 5, 5, 5, 1, 2]
```

```python
print(check(l1))
```

```python
l2=[19, 15, 5, 7, 5, 5, 2]
```

```python
print(check(l2))
```

```python
print(check(l3))
```
