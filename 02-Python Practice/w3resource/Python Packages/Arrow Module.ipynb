```python
from datetime import datetime
from dateutil import tz

d1 = arrow.get(datetime(2018, 7, 5), 'Asia/Kolkata')
print(d1)
d2 = arrow.get(datetime(2017, 7, 5), tz.gettz('America/Chicago'))
print(d2)
d3 = arrow.get(datetime.now(tz.gettz('US/Pacific')))
print(d3)
```
