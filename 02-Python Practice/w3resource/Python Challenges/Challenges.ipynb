```python
def check_pow_2(num):
    if num<=0:
        return False
    return num > 0 and (num & (num - 1)) == 0

check_pow_2(2040)
```

```python
def check_pow_3(num):
    if num<=0:
        return False
    while (num%3==0):
        num=num//3
    return num==1

check_pow_3(6)
```

```python
def check_pow(num,base):
    if num<=0:
        return False
    while (num%base==0):
        num=num//base
    return num==1

check_pow(27,3)
```

```python
def is_perfect_square(num):
    if num%(num**0.5)==0:
        return True
    return False

is_perfect_square(16)
```
