```python
import sqlite3
try:
    sqlite_con=sqlite3.connect('temp.db')
    conn=sqlite_con.cursor()
    print("Database created and connected to SQLite")
    select_query="select sqlite_version();"
    conn.execute(select_query)
    record=conn.fetchall()
    print("SQLite Database Version is: ",record)
    conn.close()
except sqlite3.Error as error:
    print("Error while connecting to sqlite", error)
finally:
    if(sqlite_con):
        sqlite_con.close()
        print("The SQLite connection is closed")
```

```python
try:
    sqlite_con=sqlite3.connect('temp.db')
    conn=sqlite3.connect(':memory:')
    print("Memory database created and connected to SQLite")
    select_query="select sqlite_version();"
    conn.execute(select_query)
    print("SQLite Database Version is: ",sqlite3.version)
    conn.close()
except sqlite3.Error as error:
    print("Error while connecting to sqlite", error)
finally:
    if(sqlite_con):
        sqlite_con.close()
        print("The SQLite connection is closed")
```
