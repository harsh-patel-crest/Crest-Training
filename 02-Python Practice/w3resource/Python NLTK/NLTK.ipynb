```python
import nltk
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import sent_tokenize
```

```python
sentence='Hiii! I am Harsh Patel. A Software Engineer. Living in Surat. Working at Crest Infosystems'
```

```python
tokens=sent_tokenize(sentence)
```

```python
for s in tokens:
    print(s)
```

```python
string='Hiii! I am Harsh Patel. A Software Engineer. Living in Surat. Working at Crest Infosystems'
```

```python
from nltk.tokenize import word_tokenize
```

```python
word_tokens=word_tokenize(string)
print(word_tokens)
```
