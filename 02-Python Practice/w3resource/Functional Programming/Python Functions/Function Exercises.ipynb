```python
def maximum(*args):
    max=args[0]
    for arg in args:
        if max<arg:
            max=arg
    return max
```

```python
maximum(1,2,2003,4,6,100,12,1,-99)
```

```python
def sum_list(l):
    result=0
    for num in l:
        result+=num
    return result
```

```python
l=[8,2,3,0,7]
sum_list(l)
```

```python
def mul_list(l):
    result=1
    for num in l:
        result*=num
    return result
```

```python
l=[8,2,3,7,-1]
mul_list(l)
```

```python
def reverse_string(s):
    return s[::-1]
```
