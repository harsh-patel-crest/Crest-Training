```python
multiply=lambda x,y: x*y
add_15=lambda x:x+15
```

```python
print(add_15(5))
print(multiply(5,10))
```

```python
def mul_num(n):
    return lambda x:x*n

result=mul_num(9)
print(result(5))
```

```python
nums=[1,2,3,4,5,6,7,8,9]
```

```python
even=list(filter(lambda x:x%2==0,nums))
odd=list(filter(lambda x:x%2!=0,nums))
```

```python
print(even)
print(odd)
```
