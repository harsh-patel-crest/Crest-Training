```python
l=[1,2,3,4,5,6,7,8,9]
def filter_even_num(num):
    def is_odd(num):
        return num%2!=0

    odd_num=list(filter(is_odd,num))
    return odd_num
```

```python
strings = ["Harsh", "w3resource", "Python", "Filter", "Crest"]
uppercase = list(filter(lambda char: char.isupper(), ''.join(strings)))
print(uppercase)
```

```python
def filter_name(l):
    l1=[]
    vowels=['A','E','I','O','U']
    for i in l:
        if i[0] in vowels:
            l1.append(i)
    return l1
```

```python
names=['Harsh','Aayushi','Akshat','Eti','Isha','Piyush']
filter_name(names)
```

```python
def starts_with_vowel(name):
    return name[0].lower() in ['a', 'e', 'i', 'o', 'u']

vowel_names = list(filter(starts_with_vowel, names))
print(vowel_names)
```
