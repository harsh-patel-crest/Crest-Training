```python
from collections import Counter

s = 'harshsunilpatel'

print("Original string: " + s)

print("Most common three characters of the string:")

print(Counter(s).most_common(3))
```

```python
from collections import Counter
import re

s = 'Python is best, best is Python, Good Job, Well Done, Done well, Well is Empty'

print("Original string: " + s)

print("Most common ten words of the string:")

words = re.findall('\w+', s)
print(Counter(words).most_common(10))
```
