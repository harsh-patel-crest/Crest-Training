```python
import json
json_obj =  '{ "Name":"Harsh", "Gender":"Male", "Age":21 }'
python_obj = json.loads(json_obj)
print(json_obj)
print(python_obj)
```

```python
print("\nJSON data:")
print(python_obj)
print("\nName: ",python_obj["Name"])
print("Gender: ",python_obj["Gender"])
print("Age: ",python_obj["Age"])
```
