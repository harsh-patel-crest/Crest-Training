```python
import seaborn as sns
data = sns.load_dataset('iris')
```

```python
print("Shape of the data:")
print(data.shape)
print("\nData Type:")
print(type(data))
print("\nFirst 3 rows:")
print(data.head(3))
```

```python
print("\nKeys of Iris dataset:")
print(data.keys())
print("\nNumber of rows and columns of Iris dataset:")
print(data.shape) 
```

```python
print(data.info())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data.describe().plot(kind = "area",fontsize=16, figsize = (15,8), table = True, colormap="Accent")
plt.xlabel('Statistics',)
plt.ylabel('Value')
plt.title("General Statistics of Iris Dataset")
plt.show()
```

```python
ax=plt.subplots(1,1,figsize=(10,8))
sns.countplot(x='species',data=data)
plt.title("Iris Species Count")
plt.show()
```

```python
ax=plt.subplots(1,1,figsize=(10,8))
data['species'].value_counts().plot.pie(explode=[0.1,0.1,0.1],autopct='%1.1f%%',shadow=True,figsize=(10,8))
plt.title("Iris Species %")
plt.show()
```
